```python
import pandas as pd
import numpy as np
import panel as pn
pn.extension('tabulator')

import hvplot.pandas
```

```python
df = pd.read_csv("energy_data.csv")
df
```

```python
df = df.fillna(0)
df.rename(columns = {"Entity":"country", "Year":"year"}, 
          inplace = True)
df
```

```python
idf = df.interactive()
```

```python
min_value = df['year'].min()
max_value = df['year'].max()

year_slider = pn.widgets.IntSlider(name = "YEAR", 
                                   start = min_value, 
                                   end = max_value, 
                                   step = 1, 
                                   value = min_value)
year_slider
```

```python
df.columns
```

```python
yaxis_access = pn.widgets.RadioButtonGroup(
    name = "Y AXIS", 
    options = ["Access to electricity (% of population)", "Access to clean fuels for cooking",],
    button_type = "success"
)

biggest_countries = ["Canada", "United States", "China", "Brazil", "Australia"]

access_pipeline = (
    idf[
        (idf.year <= year_slider) &
        (idf.country.isin(biggest_countries))
    ]
    .groupby(["country", "year"])[yaxis_access].mean()
    .to_frame()
    .reset_index()
    .sort_values(by = "year")  
    .reset_index(drop=True)
)
```

```python
access_pipeline
```

```python
access_plot = access_pipeline.hvplot(x = "year", 
                                     by = "country", 
                                     y = yaxis_access, 
                                     line_width = 2, 
                                     title = "Accessibility Percentages in Top 5 Biggest Countries")
access_plot
```

```python
access_table = access_pipeline.pipe(pn.widgets.Tabulator, pagination = "remote", page_size = 10, sizing_mode = "stretch_width") 
access_table
```

```python
finance_vs_ren_scatterplot_pipeline = (
    idf[
        (idf.year == year_slider) &
        (~ (idf.country.isin(biggest_countries)))
    ]
    .groupby(["country", "year", "Access to clean fuels for cooking"])["Renewable energy share in the total final energy consumption (%)"].mean()
    .to_frame()
    .reset_index()
    .sort_values(by = "year")  
    .reset_index(drop = True)
)
```

```python
finance_vs_ren_scatterplot = finance_vs_ren_scatterplot_pipeline.hvplot(x = "Access to clean fuels for cooking", 
                                                                        y = "Renewable energy share in the total final energy consumption (%)", 
                                                                        by = "country", 
                                                                        size = 80, 
                                                                        kind = "scatter", 
                                                                        alpha = 0.7,
                                                                        legend = False, 
                                                                        height = 500, 
                                                                        width = 500
                                                                       )
finance_vs_ren_scatterplot
```

```python
yaxis_usage = pn.widgets.RadioButtonGroup(
    name = "Y AXIS", 
    options=["Value_co2_emissions_kt_by_country", "Electricity from renewables (TWh)", "Energy intensity level of primary energy (MJ/$2017 PPP GDP)"], 
    button_type = "success"
)

usage_bar_pipeline = (
    idf[
        (idf.year == year_slider) &
        (idf.country.isin(biggest_countries))
    ]
    .groupby(["year", "country"])[yaxis_usage].sum()
    .to_frame()
    .reset_index()
    .sort_values(by = "year")  
    .reset_index(drop = True)
)
```

```python
co2_usage_plot = usage_bar_pipeline.hvplot(kind = "bar", 
                                           x = "country", 
                                           y = yaxis_usage, 
                                           title = "Usages in Top 5 Biggest Countries")
co2_usage_plot
```

```python
import panel.widgets as pnw



template = pn.template.FastListTemplate(
    title='Top 5 Biggest Countries Statistics on Accessibility and Resource Usages', 
    sidebar=[pn.pane.Markdown("## Problem at Hand: "), 
             pn.pane.Markdown("#### Global energy consumption patterns are problematic due to their reliance on finite resources, environmental degradation from fossil fuel usage, and unequal access, exacerbating socioeconomic disparities worldwide."), 
             pn.pane.Markdown("## Background on Dataset"), 
             pn.pane.Markdown("#### The purpose of this dataset was to explore Global energy consumption patterns and indicators from all nations provided"), 
             pn.pane.Markdown("## Adjustments: "),   
             year_slider
            ],
    main=[pn.Row(pn.Column(yaxis_access, 
                           access_plot.panel(width = 700), 
                           margin = (0, 25)), 
                 
                 access_table.panel(width = 500)), 
          
          pn.Row(pn.Column(finance_vs_ren_scatterplot.panel(width = 600), 
                           margin = (0, 25)), 
                 
                 pn.Column(yaxis_usage, 
                           co2_usage_plot.panel(width = 600)))],
    accent_base_color = "#5797ff",
    header_background = "#5797ff",
)

template.show()
template.servable();
```

```python
import panel.widgets as pnw

def save_dashboard(event, column):
  column.save(filename="EnergyData.html" , title="Data on Accessibility and Resource Usages")

full_dashboard = pn.Column(pn.Row("# Top 5 Biggest Countries Statistics on Accessibility and Resource Usages"), 
                           pn.Row("## Problem at Hand"), 
                           pn.Row("#### Global energy consumption patterns are problematic due to their reliance on finite resources, environmental degradation from fossil fuel usage, and unequal access, exacerbating socioeconomic disparities worldwide."), 
                           pn.Row("## Background on Dataset"), 
                           pn.Row("#### The purpose of this dataset was to explore Global energy consumption patterns and indicators from all nations provided."), 
                           pn.Row(pn.Column(access_plot), 
                                  pn.Column(access_table)
                                 ), 
                           pn.Row(pn.Column(finance_vs_ren_scatterplot), 
                                  pn.Column(co2_usage_plot)
                                 )
                          )

save_button = pnw.Button(name="Save as .html", button_type="primary")
save_function = pn.bind(save_dashboard, column = full_dashboard)
save_button.on_click(save_function)
full_dashboard.append(save_button)

full_dashboard
```
